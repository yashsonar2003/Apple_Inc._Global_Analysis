# EDA (Exploratory Data Analysis)

---

Objectives:-
- To analyze revenue trends of Apple Inc. over time to understand growth patterns.
- To examine profit margin distribution across products and regions.
- To compare product-wise sales performance globally.
- To evaluate regional contribution to total revenue and profitability.
- To understand the impact of marketing spend on revenue generation.
- To explore stock price trends in relation to business performance.
- To identify relationships between key business variables using correlation analysis.

---

Import Libraries & Load Processed Data:-

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('../Data/03_Processed_Data/Apple_Global_Processed_Data.csv')
```

- Convert date column


```python
df['date'] = pd.to_datetime(df['date'])
df.head()
```

Revenue Trends Over Time:-

```python
Revenue_Trend = df.groupby('date')['revenue_usd_million'].sum().reset_index()

plt.figure(figsize = (12, 6))
plt.plot(Revenue_Trend['date'], Revenue_Trend['revenue_usd_million'])
plt.xlabel('Date')
plt.ylabel('Revenue (USD Million)')
plt.title('Revenue Trend Over Time - Apple Inc.')
plt.grid(True)
plt.show()
```

Profit Margin Distribution:-

```python
plt.figure(figsize = (8, 5))
plt.hist(df['profit_margin_%'], bins = 30)
plt.xlabel('Profit Margin %')
plt.ylabel('Frequency')
plt.title('Profit Margin Distribution')
plt.show()
```

Product-wise Sales Comparison:-

```python
Product_Sales = df.groupby('product_category')['total_sales_value_usd'].sum().sort_values(ascending = False)

plt.figure(figsize = (10,6))
Product_Sales.plot(kind = 'bar')
plt.xlabel('Product Category')
plt.ylabel('Total Sales Value (USD)')
plt.title('Product-wise Sales Performance')
plt.xticks(rotation = 0)
plt.grid(axis = 'y')
plt.show()
```

Region-wise Revenue Performance:-

```python
Region_Revenue = df.groupby('region')['revenue_usd_million'].sum().sort_values(ascending=False)

plt.figure(figsize = (12, 6))
Region_Revenue.plot(kind = 'bar')
plt.xlabel('Region')
plt.ylabel('Revenue (USD Million)')
plt.title('Region-wise Revenue Performance')
plt.xticks(rotation = 0)
plt.grid(axis = 'y')
plt.show()
```

Marketing Spend vs Revenue:-

```python
plt.figure(figsize = (8,6))
plt.scatter(df['marketing_spend_usd_million'], df['revenue_usd_million'])
plt.xlabel('Marketing Spend (USD Million)')
plt.ylabel('Revenue (USD Million)')
plt.title('Marketing Spend vs Revenue')
plt.xticks(rotation = 0)
plt.show()
```

Yearly Average Apple Stock Closing Price Trend:-

```python
Yearly_Stock = df.groupby('Year')['stock_close_usd'].mean().reset_index()

plt.figure(figsize=(10,5))
plt.plot(Yearly_Stock['Year'], Yearly_Stock['stock_close_usd'])
plt.xlabel("Year")
plt.ylabel("Average Stock Close Price (USD)")
plt.title("Yearly Average Apple Stock Closing Price Trend")
plt.show()
```

Correlation Heatmap:-

```python
Numeric_Cols = [
    'units_sold','revenue_usd_million','average_price_usd',
    'customer_rating','return_rate_%','online_sales_%',
    'retail_sales_%','stock_open_usd','stock_close_usd',
    'stock_high_usd','stock_low_usd',
    'marketing_spend_usd_million','profit_margin_%',
    'total_sales_value_usd','profit_amount_usd_million'
]

corr = df[Numeric_Cols].corr()

plt.figure(figsize=(14,10))
sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap - Business Metrics")
plt.show()
```

---
Key Insights:-
- Revenue of Apple Inc. shows a consistent upward trend over time, indicating strong global growth and increasing market demand across years.
- Profit margin distribution is mostly concentrated in the medium to higher range, suggesting that Apple maintains healthy profitability across its products and regions.
- Product-wise analysis highlights that certain categories (such as iPhone and Services) contribute significantly higher total sales value compared to other products.
- Region-wise revenue comparison shows that a few regions dominate overall revenue generation, while some regions contribute relatively lower, indicating potential growth opportunities.
- There is a positive relationship between marketing spend and revenue, implying that increased marketing investments generally lead to higher revenue generation.
- Stock closing price trends follow an overall increasing pattern, reflecting long-term business growth and positive investor confidence aligned with company performance.
